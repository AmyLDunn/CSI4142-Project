```python
import pandas as pd
```

```python
#Create a dataframe
df = pd.DataFrame(columns = ['demographics_key', 'dissemination_area', 'population', 'average_age', 'median_age', 'total_dwellings', 'average_household_size', 'median_household_income','mother_tongue_official_percentage','mother_tongue_unofficial_percentage','census_year'])
df.head()
```

```python
#Create variables that will be stored in database
year = 0
key = ''
dissemination_area =0
population =0
avg_age =0
med_age =0
total_dwellings =0
avg_household_size =0
med_household_income=0
total_mother_tongue=0
total_unofficial=0
mother_tongue_official_per=0.0
mother_tongue_unofficial_per=0.0

#insertion index
index=0

#functions to handle null values
def convert_to_int(val,row):
    try:
        return int(val)
    except:
        print(row)
        return ''
def convert_to_float(val,row):
    try:
        return float(val)
    except:
        print(row)
        return ''
def calculate_unofficial(val1,val2):
    if (val1==''or val2==''):
        return ''
    elif(val2==0):
        return 0
    else:
        return round((100*(val1/val2)),2)

def calculate_official(val):
    if val=='':
        return ''
    else:
        return 100-val
#removes the double quotes from the strings (used for key and dissemination_area)
def remove_quotes(val):
    return val.replace('"','')

#Open file and read line by line and append to
with open('98-401-X2016044_ONTARIO_English_CSV_data.csv','r') as file_2016:
    for line in file_2016:
#         (CENSUS_YEAR,DGUID, _ ,GEO_LEVEL, GEO_NAME, _, _, _, CHARACTERISTIC_ID, _, C1_Count_Total,_,_,_,_,_,_,_,_,_,_,_ )= 
        temp_data = line.split(',')
#        if index==10:
#            break
        if temp_data[2]=='"4"':
            if temp_data[9]=='39':
                avg_age=convert_to_float(temp_data[11],temp_data)
            elif temp_data[9]=='40':
                med_age=convert_to_float(temp_data[11],temp_data)
            elif temp_data[9]=='41':
                total_dwellings=convert_to_int(temp_data[11],temp_data)
            elif temp_data[9]=='58':
                avg_household_size=convert_to_float(temp_data[11],temp_data)
            elif temp_data[9]=='663':
                med_household_income=convert_to_float(temp_data[11],temp_data)
            elif temp_data[9]=='105':
                total_mother_tongue=convert_to_int(temp_data[11],temp_data)
            elif temp_data[9]=='109':
                total_unofficial=convert_to_int(temp_data[11],temp_data)
            elif temp_data[9]=='664':
                #calculate mother tongue statistics
                mother_tongue_unofficial_per = calculate_unofficial(total_unofficial,total_mother_tongue)
                mother_tongue_official_per = calculate_official(mother_tongue_unofficial_per)
                #create a row in a dataframe
                df.loc[index]=[key, dissemination_area, population, avg_age, med_age, total_dwellings, avg_household_size, 
                                        med_household_income, mother_tongue_official_per, mother_tongue_unofficial_per, year]
                #increment index
                index = index+1
            elif temp_data[10]=='1':
                year=temp_data[0]
                dissemination_area=remove_quotes(temp_data[1])
                key=year+dissemination_area
                year=convert_to_int(year,temp_data)
                dissemination_area=convert_to_int(dissemination_area,temp_data)
                key=convert_to_int(key,temp_data)
                population = convert_to_int(temp_data[12],temp_data)
file_2016.close()        
```
