```python
import pandas as pd
```

```python
#Create a dataframe
df = pd.DataFrame(columns = ['demographics_key', 'dissemination_area', 'population', 'average_age', 'median_age', 'total_dwellings', 'average_household_size', 'median_household_income','mother_tongue_official_percentage','mother_tongue_unofficial_percentage','census_year'])
df.head()
```

```python
#Create variables that will be stored in database
year = 0
key = ''
dissemination_area =0
population =0
avg_age =0
med_age =0
total_dwellings =0
avg_household_size =0
med_household_income=0
total_mother_tongue=0
total_unofficial_single=0
total_unofficial_mult=0
mother_tongue_official_per=0.0
mother_tongue_unofficial_per=0.0

#insertion index
index=0

#functions to handle null values
def convert_to_int(val):
    if val=='':
        return ''
    else:
        return int(val)
def convert_to_float(val):
    if val=='':
        return ''
    else:
        return float(val)
def calculate_unofficial(val1,val2,val3):
    if (val1==''or val2=='' or val3==''):
        return ''
    elif(val3==0):
        return 0
    else:
        return round((100*((val1+val2)/val3)),2)

def calculate_official(val):
    if val=='':
        return ''
    else:
        return 100-val
#removes the double quotes from the strings (used for key and dissemination_area)
def remove_quotes(val):
    return val.replace('"','')

def simplify_key(string,value_to_remove):
    return string.replace(value_to_remove,'')

#Open file and read line by line and append to
with open('98-401-X2021006_English_CSV_data_Ontario.csv','r') as file_2021:
    for line in file_2021:
#         (CENSUS_YEAR,DGUID, _ ,GEO_LEVEL, GEO_NAME, _, _, _, CHARACTERISTIC_ID, _, C1_Count_Total,_,_,_,_,_,_,_,_,_,_,_ )= 
        temp_data = line.split(',')
#        if index==30:
#            break
        if temp_data[3]=='"Dissemination area"':
            if temp_data[8]=='1':
                year=convert_to_int(temp_data[0])
                key=remove_quotes(simplify_key(temp_data[1],'S0512'))
                dissemination_area=convert_to_int(remove_quotes(temp_data[4]))
                population = convert_to_int(temp_data[12])
            elif temp_data[8]=='39':
                avg_age=convert_to_float(temp_data[11])
            elif temp_data[8]=='40':
                med_age=convert_to_float(temp_data[11])
            elif temp_data[8]=='41':
                total_dwellings=convert_to_int(temp_data[11])
            elif temp_data[8]=='57':
                avg_household_size=convert_to_float(temp_data[11])
            elif temp_data[8]=='243':
                med_household_income=convert_to_float(temp_data[11])
            elif temp_data[8]=='393':
                total_mother_tongue=convert_to_int(temp_data[11])
            elif temp_data[8]=='398':
                total_unofficial_single=convert_to_int(temp_data[11])
            elif temp_data[8]=='723':
                total_unofficial_mult=convert_to_int(temp_data[11])
            elif temp_data[8]=='724':
                #calculate mother tongue statistics
                mother_tongue_unofficial_per = calculate_unofficial(total_unofficial_single,total_unofficial_mult,total_mother_tongue)
                mother_tongue_official_per = calculate_official(mother_tongue_unofficial_per)
                #create a row in a dataframe
                df.loc[index]=[key, dissemination_area, population, avg_age, med_age, total_dwellings, avg_household_size, 
                                        med_household_income, mother_tongue_official_per, mother_tongue_unofficial_per, year]
                #increment index
                index = index+1
file_2021.close()        
```
